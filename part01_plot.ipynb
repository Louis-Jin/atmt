### Calculate BLEU scores and plot

```python
ks=[1,2,3,5,7,8,9,10,11,20,30,50]
import sacrebleu
import numpy as np
bleus = []
for i in ks:
    with open('data_asg4/raw_data/test.en', 'r') as f:
        refs = f.readlines()
    refs = [refs]
    with open('./output_asg4_part4/k{}/postprocessed/model_translations_alpha0-gamma0.out'.format(i), 'r') as f:
        sys = f.readlines()
    bleu = sacrebleu.corpus_bleu(sys, refs)
    bleus.append(bleu.score)
    
import matplotlib.pyplot as plt
plt.plot(ks, bleus, marker='o')
plt.xlabel('ks')
plt.ylabel('BLEU')
plt.title('Beam Search Size Experiment')
```

### plot average length of prediction 

```python
ks=[1,2,3,5,7,8,9,10,11,20,30,50]
avg_lens = []
for i in ks:
    with open('./output_asg4_part4/k{}/postprocessed/model_translations_alpha0-gamma0.out'.format(i), 'r') as f:
        sys = f.readlines()
    avg_len = 0
    for j in range(0, len(sys)):
        avg_len += len(sys[j])
    avg_lens.append(avg_len/len(sys))
    
plt.plot(ks, avg_lens, marker='o')
plt.xlabel('ks')
plt.ylabel('avg translation len')
plt.ylim(22, 33)

plt.title('Beam Search Size Experiment')
```
